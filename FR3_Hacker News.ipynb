# Technical Approach

#### Step 1 - Read the CSV file and create a dictionary. The Key will be the (Post Type + Hour) and data will be a list containing [Points,Count of Points(same as count of posts),Comments,Count of comment]. While loading the dictionary since we are reading the complete CSV file, so we will count the number of Ask HN and Show HN posts. This will address Bullet 1.

#### Step 2 - Read the dictionary sequenctially.  Sum up ListData[0] ,ListData[1] ListData[,2] ListData[3] into two group of variables. Group A will be for Ask HN related Keys and Group B will be Show HN relates Keys. Address Bullet 2,3,8,9 and 10.
-  Avg Points Awarded  =  Sum ListData[0] / Sum ListData[1]
-  Avg Comments Posted  =  Sum ListData[2] / Sum ListData[3]
-  Compare Avg. Points - Address Bullet 4
-  Compare Avg. Comments - Address Bullet 5
-  While reading dictionary keep comparing Listdata[1] and save the max, along with the Key. This tells the hour at which max points are awarded. (Bullet 9)
-  While reading dictionary keep comparing Listdata[2] and save the max, along with the Key. This tells the hour at which max comments are provided. (Bullet 8)

#### Step 2.1 - Since we are reading the complete dictionary , create 4 lists as below , which will be useful to determine the top hours when  posts recieve max comments or max points are awarded.Address bullet 6 and 7. We are creating below 4 lists from the dictionary.
   ` post_by_hour_ask`
    `points_by_hour_ask`
   ` post_by_hour_show`
    `points_by_hour_show`


-  Divide the ListData[0] with ListData[1] to get the Average Number of Points for Ask Posts and Show Posts (Bullet 2)
-  Divide the ListData[2] with ListData[3] to get the Average Number of Comments for Ask Points and Show Posts (Bullet 3)
-  While reading dictionary keep comparing Listdata[1] and save the max, along with the Key. This tells the hour at which max points are awarded. 
-  While reading dictionary keep comparing Listdata[2] and save the max, along with the Key. This tells the hour at which max comments are provided. 

```python
# Step 1 - Read the file

from csv import reader 
import datetime as dt
from collections import OrderedDict

opened_file = open(r"C:\Users\DELL\Desktop\DGD\Project 1\hacker_news.csv")
read_file = reader(opened_file) # Create a reader object so we can iterate over rows of CSV file.
hn = list(read_file)            # dump the reader object into a list.
headers = hn[:1]                # Extract the header part into a variable 'headers'. it is a list
hn = hn[1:]                     # Extarct the data part (non header) into a variable called 'hn'. It is a List.
```

```python
# Step 1.1 - Create a dictinary.
#     Key   :   [Sum of Points,  Count of Points/posts  ,  Sum of Comments , Count comments/posts  ]    
#[('ask_hn00', [451.0         ,      55.0               ,       447.0      ,          55.0   ]), ('ask_hn01', [700.0, 60.0, 683.0, 60.0]).........]
  

dict1={}
list1=[]
post_by_hour_ask = []
post_by_hour_show = []

ask_hn_comments = 0
ask_hn_posts = 0
nbr_of_ask_post = 0
nbr_of_show_post = 0

for csvdata in hn:
    proceed = 'N'
    title=csvdata[1]
    id1=csvdata[0]
    points= float(csvdata[3])
    comments=float(csvdata[4])
    date_str=csvdata[6]
    dt_object = dt.datetime.strptime(date_str, '%m/%d/%Y %H:%M')
    hour = dt_object.strftime('%H')
        
    if title.lower().startswith("ask hn"):
        key="ask_hn" + hour
        nbr_of_ask_post= nbr_of_ask_post+1
        proceed = 'Y'
    elif title.lower().startswith("show hn"):
        key="show_hn" + hour
        nbr_of_show_post = nbr_of_show_post + 1
        proceed = 'Y'

    if proceed == 'Y':
        if dict1.get(key) == None:
            dict1[key]=[float(points),1.0,float(comments),1.0]
        else:
            dict_points = dict1[key][0] + points
            dict_point_count = dict1[key][1] + 1
            dict_comments = dict1[key][2] + comments
            dict_comments_count = dict1[key][3] + 1
            dict1[key]=[dict_points,dict_point_count,dict_comments,dict_comments_count]

print('Numbe of ask post:',nbr_of_ask_post)
print('Number of show post:', nbr_of_show_post)
#list2=list(sorted(dict1.items()))   # For testing purpose only.....Put Dictionary into list.
for i,j in sorted(dict1.items()):
    print(i,j)
```

```python
#Step 2 - Read the dictinary and determine time of max comments on posts , max point awarded on posts. Also create 2 lists which 
# can be used to sort the hourly comments and hourly points awarded.

#Average Post comments
     # Read the dictionay with Key as 'ask_hnxx' or !'ask_hnn' (i.e Show_hnxx)
sum_points_ask = 0
sum_comments_ask = 0
sum_points_show = 0
sum_comments_show = 0
sum_point_ask_count = 0
sum_comments_ask_count = 0
sum_points_show_count = 0
sum_comments_show_count = 0
avg_points_ask = 0
avg_comments_ask = 0
avg_points_show = 0
avg_comments_show = 0
save_max_posts_nbr_ask = 0
save_max_posts_nbr_show = 0
save_max_comments_ask = 0
save_maxcommenttime_ask = 0
save_max_comments_show = 0
save_maxcommenttime_show = 0
save_max_points_ask = 0
save_maxpointtime_ask = 0
save_max_points_show = 0
save_maxpointtime_show = 0
post_by_hour_ask = []
post_by_hour_show = []
post_by_hour_all =[]
more_comments_rcvd = ''
more_points_rcvd = ''
points_by_hour_ask = []
points_by_hour_show = []

for key,data in sorted(dict1.items()):
    if key[:6] == 'ask_hn':
        post_by_hour_ask.append((data[2],'Hour '+ key[6:8]))   # Comments Rcvd by Ask Posts by hour
        points_by_hour_ask.append((data[0],'Hour '+ key[6:8]))   # Points Rcvd by Ask Posts by hour
        sum_points_ask = sum_points_ask + data[0]                  # Bullet 2
        sum_point_ask_count = sum_point_ask_count + data[1]
        sum_comments_ask = sum_comments_ask + data[2]             # Bullet 3
        sum_comments_ask_count = sum_comments_ask_count + data[3]
        if data[0] >= save_max_points_ask:  # Max points in which hour ? # Bullet 10
            save_max_points_ask = data[0]
            save_maxpointtime_ask = key[6:]
        if data[1] >= save_max_posts_nbr_ask:  # Max posts in which hour ?  # Bullet 8
            save_max_posts_nbr_ask = data[1]
            save_maxposttime_ask = key[6:]
        if data[2] >= save_max_comments_ask:  #  Max comments in which hour ? # Bullet 9
            save_max_comments_ask = data[2]
            save_maxcommenttime_ask = key[6:]                    
    else:
        post_by_hour_show.append((data[2],'Hour '+ key[7:9])) 
        points_by_hour_show.append((data[0],'Hour '+ key[7:9]))   # Points Rcvd by Ask Posts by hour
        sum_points_show = sum_points_show + data[0]
        sum_points_show_count = sum_points_show_count + data[1]
        sum_comments_show = sum_comments_show + data[2]
        sum_comments_show_count = sum_comments_show_count + data[3]
        if data[0] >= save_max_points_show:            # Bullet 10
            save_max_points_show = data[0]
            save_maxpointtime_show = key[7:]
        if data[1] >= save_max_posts_nbr_show:        # Bullet 8
            save_max_posts_nbr_show = data[1]
            save_maxposttime_show = key[7:]
        if data[2] >= save_max_comments_show:        # Bullet 9
            save_max_comments_show = data[2]
            save_maxcommenttime_show = key[7:]  
    
avg_points_ask = sum_points_ask/sum_point_ask_count                # Bullet 3
avg_comments_ask = sum_comments_ask/sum_comments_ask_count         # Bullet 2
avg_points_show = sum_points_show /sum_points_show_count           # Bullet 2
avg_comments_show = sum_comments_show/sum_comments_show_count      # Bullet 3

##########################################################################################################################
def print_stats(post_category,avg_points,avg_comments,save_maxpointtime,save_max_points,save_maxposttime,save_max_posts_nbr,save_maxcommenttime,save_max_comments):
    print("Stats Pertaining to " + post_category + " Category")
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print(f'Average number of points awarded per posts in Ask HN are: {(avg_points):.2f}')
    print(f'Average ask post Comments : {avg_comments_ask:.2f}')
    print(f'Max Points awarded to ask posts is at time {save_maxpointtime_ask} and the count is {save_max_points_ask}')
    print(f'Max ask Posts are posted at time {save_maxposttime_ask} and the count is {save_max_posts_nbr_ask}')
    print(f'Max ask Post comments are at time {save_maxcommenttime_ask} and the count is {save_max_comments_ask}')
##########################################################################################################################
    

print_stats("Ask HN",avg_points_ask,avg_comments_ask,save_maxpointtime_ask,save_max_points_ask,save_maxposttime_ask,save_max_posts_nbr_ask,save_maxcommenttime_ask,save_max_comments_ask)
print('\n')
print_stats("Show HN",avg_points_show,avg_comments_show,save_maxpointtime_show,save_max_points_show,save_maxposttime_show,save_max_posts_nbr_show,save_maxcommenttime_show,save_max_comments_show)
print('\n')


# COMMENTS COMPARISON : Which Type of Posts received more comments and how much.   Bullet 4
if avg_comments_ask > avg_comments_show:
    print(f'Ask posts recieved {(avg_comments_ask-avg_comments_show):.2f} more average comments')
    more_comments_rcvd = "Ask HN"
elif avg_comments_show > avg_comments_ask:
    print(f'Show posts recieved more average comments')
    more_comments_rcvd = "Show HN"
else:
    print(f'Ask posts and Show posts recieved same nunmber of average comments')

# POINTS COMPARISON : Which Type of Posts received more points and how much.    # Bullet 5
if avg_points_ask > avg_points_show:
    print(f'Ask posts recieved {(avg_points_ask-avg_points_show):.2f} more average points')
    more_points_rcvd = "Ask HN"
elif avg_points_show > avg_points_ask:
    print(f'Show posts recieved {(avg_points_show-avg_points_ask):.2f} more average points')
    more_points_rcvd = "Show HN"
else:
    print(f'Ask posts and Show posts recieved same nunmber of average comments')

print('\n')

##########################################################################################################################
# Ideal time to Post to get Max comments dict1{2} = Nbr of aggregated comments for a Post  dict1{3}= Count of posts during an hour
def best_time_to_post_for_max(post_by_hour,str):
    sorted_post_by_hour = sorted(post_by_hour)
    str=('')
    count=1
    temp_str=','
    for i,j in sorted_post_by_hour[-1:-6:-1]:
        str=str + j[5:7]
        if count==4:
            str=str+ ' or '
        elif count <5:
            str=str+','
        count=count + 1
    return(str)

x = best_time_to_post_for_max(post_by_hour_ask,str)
print('For Ask Posts to get most comments the best time is:',x)      # Bullet 6
x = best_time_to_post_for_max(post_by_hour_show,str)
print('For Show Posts to get most comments the best time is:',x)     # Bullet 6
x = best_time_to_post_for_max(points_by_hour_ask,str)  
print('For Ask Posts to get most points the best time is::',x)       # Bullet 7
x = best_time_to_post_for_max(points_by_hour_show,str)
print('For Show Posts to get most points the best time is:',x)       # Bullet 7
##########################################################################################################################  



    
    
print('\n')
print('Ask Posts by Hour for comments ')
print(sorted(post_by_hour_ask))
print('\n')
print('Show Posts by Hour for comments ')
print(sorted(post_by_hour_show))
print('\n')
print(sorted(dict1.items()))
```
